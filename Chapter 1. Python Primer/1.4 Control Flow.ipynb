1.4 Control Flow
=====
***
* The colon character is used to delimit the beginning of a block of code that acts as a body for a control structure.
* Python relies on the identation level to designate the extent of that block of code, or any nested blocks of code within.

1.4.1 Conditionals
-----
***
In Python, the most general form of a conditional is written follows:
> __if__ _first condition_:<br>
>&nbsp;&nbsp;&nbsp;&nbsp;_first body_<br>
>__elif__ _second condition_:<br>
>&nbsp;&nbsp;&nbsp;&nbsp;_second body_<br>
>__elif__ _second condition_:<br>
>&nbsp;&nbsp;&nbsp;&nbsp;_second body_<br>
>__elif__ _third condition_:<br>
>&nbsp;&nbsp;&nbsp;&nbsp;_third body_<br>
> __else__:<br>
>&nbsp;&nbsp;&nbsp;&nbsp;_fourth body_

* The execution of this overall construct will cause precisely one of the bodies to be executed.

1.4.2 Loops
-----
***
* A __while__ loop allows general repetition based upon the repeated testing of a Boolean condition.
* A __for__ loop provides convenient iteration of values from a defined series, such as characters of a string, elements of a list, or numbers within a given range.

### While Loops
* As an example, here is a loop that advances an index through a sequence of characters until finding an entry with value 'X' or reaching the end of the sequence.
    * The `len` function returns the length of sequence such as a list or string.
    * Had we written that compound condition with the opposite order, the evaluation of data[j] would eventually raise an `IndexError` when 'X' is not found.
    * This code behaves correctly, even in the special case when the list is empty, as the condition `j < len(data)` will initially fail and the body of the loop will never be executed.

> j = 0<br>
> __while__ j < len(data) __and__ data[j] != 'X':<br>
> &nbsp;&nbsp;&nbsp;&nbsp;j += 1

### For Loops
* The for-loop syntax can be used on any type of __iterable__ structure, such as a `list`, `tuple`, `str`, `set`, `dict`, or `file`.
* There is no need to manage an explicit index into the list nor th author a Boolean loop condition.
* We can use a for loop in cases for which a while loop does not apply, such as when iterating through a collection, such as a `set`, that does not support any direct form of indexing.
    * The loop body executes once for each element of the data sequence, with the identifier, `val`, from the for-loop syntax assigned at the beginning of each pass to a respective element.
    * `val` is treated as a standard identifier. But a reassignment of identifier `val` to a new value has no affect on the original data, nor on the next iteration of the loop.
> total = 0<br>
> __for__ val __in__ data:<br>
> &nbsp;&nbsp;&nbsp;&nbsp;total += val

### Index-Based For Loops
* In some applications, we need knowledge of the index of an element within the sequence.
* We prefer to loop over all possible indices of the list. For this purpose, Python provides a built-in class named `range` that generates integer sequences. (`n` values from `0` to `n-1`)
    * For examle, we can find the _index_ of the maximum element of a list as follows:
> bing_index = 0<br>
> __for__ j __in__ range(len(data)):<br>
> &nbsp;&nbsp;&nbsp;&nbsp;__if__ data[j] > data[big_index]:<br>
> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;big_indes = j

### Break and Continue Statements
* Python supports a __break__ statement that immediately terminate a while or for-loop when executed within its body.
* It also supports a __continue__ statement that causes the current _iteration_ to stop, but with subsequent passes as expected.

> found = __False__
> __for__ item __in__ data:<br>
